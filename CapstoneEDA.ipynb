### Variables Description (Only for Fetures used for EDA)

**EventID** - (eventid) Numeric Variable - Incidents from the GTD follow a 12-digit Event ID system.

**Year** - (iyear) Numeric Variable - This field contains the year in which the incident occurred.

**Month** - (imonth) Numeric Variable - This field contains the number of the month in which the incident occurred.

**Day** - (iday) Numeric Variable - This field contains the numeric day of the month on which the incident occurred.

**Country** - (country; country_txt) Categorical Variable -  This field identifies the country or location where the incident occurred.

**Region** - (region; region_txt) Categorical Variable - This field identifies the region in which the incident occurred.

**City** - (city) Text Variable - This field contains the name of the city, village, or town in which the incident occurred.

**AttackType** - (attacktype1; attacktype1_txt) Categorical Variable - This field captures the general method of attack and often reflects the broad class of tactics used.

**Killed** - (nkill) Numeric Variable - This field stores the number of total confirmed fatalities for the incident.

**Wounded** - (nwound) Numeric Variable - This field records the number of confirmed non-fatal injuries to both perpetrators and victims.

**Casualities** - Numeric Variable - This field records the sum of Killed and Wounded features

**Target** - (target1) Text Variable - This is the specific person, building, installation, etc., that was targeted and/or victimized and is a part of the entity named above.

**Group** - (gname) Text Variable - This field contains the name of the group that carried out the attack.

**TargetType** - (targtype1; targtype1_txt) Categorical Variable - The target/victim type field captures the general type of target/victim.

**WeaponType** - (weaptype1; weaptype1_txt) Categorical Variable - Up to four weapon types are recorded for each incident. This field records the general type of weapon used in the incident.

**Success** - (success) Categorical  Variable - Success of a terrorist strike is defined according to the tangible effects of the attack. Success is not judged in terms of the larger goals of the perpetrators.

**Suicide** - (suicide) Categorical Variable  This variable is coded “Yes” in those cases where there is evidence that the perpetrator did not intend to escape from the attack alive.
